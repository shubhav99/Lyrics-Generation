```python
#Use of Markov Chain probability theorem
# text = the man was ...... they  ...then........the...... the.......
# X=(input) , y=(output)
# k=3
X.       y.     freq
the      -       0+i
the      y       0+i
the      n       0+i
```

```python
#definig the function based on Markov Chain probability theorem
def generatetable(data, k=4):
    T = { }
    
    for i in range(len(data)-k):
        X = data[ i : i+k ]
        y = data[ i+k ]
        
        if T.get(X) is None: #if input state is not in the dictionary
            T[X] = { } #initializing empty dictionary
            T[X][y] = 1
        else:
            if T[X].get(y) is None:  # If you already have input in dictionary but the output is not in dictionary
                T[X][y] = 1
            else:
                T[X][y] +=1  # if you have both input and output in dic, just increment the frequency.
                
    return T 
```

```python
#sample data to test the function defined
d = 'hello helli hello helly helli hello hello'
a=generatetable(d, k=4)
```
